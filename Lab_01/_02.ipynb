```python
import numpy as np
import matplotlib.pyplot as plt

nmin, nmax = -5, 15
n = np.arange(nmin, nmax+1)

x = lambda n: 3*((n==0)|(n==1)) + 5*((n==2)|(n==3))+ 4*(n==4) + 3*(n==5) + 2*(n==6)+ 1*(n==7) # | = or 

fig,ax=plt.subplots(1,3,figsize=(6,2),dpi=100)
# x[-n]
ax[0].stem(n, x(-n))
ax[0].set(xlim=[-10, 1],ylim=[-0.25,6],title='x[-n]',xlabel='n')
# x[n/2]
ax[1].stem(n, x(n/2))
ax[1].set(xlim=[-1, 15],ylim=[-0.25,6],title='x[n/2]',xlabel='n')
# x[n*2]
ax[2].stem(n, x(n*2))
ax[2].set(xlim=[-1,6],ylim=[-0.25,6],title='x[n*2]',xlabel='n')
```

```python
import numpy as np
import matplotlib.pyplot as plt
n = np.arange(-5, 15+1)
# generate signals
imp = lambda n: (n==0)*(n==np.round(n)).astype(float)# impulse
u = lambda n: (n>=0)*(n==np.round(n)).astype(float)# step
x1 = np.exp(-0.5*n)*u(n)# exponential
x2 = np.cos(2*np.pi*n/12)# sinusoid
fig,bx=plt.subplots(2,2,figsize=(6,4),dpi=100)
bx[0,0].stem(n,u(n));bx[0,0].set(title='unit impluse')
bx[0,1].stem(n,imp(n));bx[0,1].set(title='unit step')
bx[1,0].stem(n,x1);bx[1,0].set(title='exponential',xlabel='n')
bx[1,1].stem(n,x2);bx[1,0].set(title='sinusoid',xlabel='n')

plt.tight_layout();
```

```python
import numpy as np
import matplotlib.pyplot as plt
nmin, nmax = -10, 10; n = np.arange(nmin, nmax+1)
u = lambda n: (n>=0).astype(float)
ueven = 0.5*(u(n) + u(-n)); uodd = 0.5*(u(n) - u(-n))
x1 = lambda n: np.exp(-0.5*n)*u(n)
x2 = lambda n: np.cos(2*np.pi*n/12) + np.sin(2*np.pi*n/12)
x1_even = 0.5*(x1(n) + x1(-n)); x1_odd = 0.5*(x1(n) - x1(-n))
x2_even = 0.5*(x2(n) + x2(-n)); x2_odd = 0.5*(x2(n) - x2(-n))
fig, axes = plt.subplots(3,3, figsize=(6,4.5), dpi=100); 
axes[0,0].stem(n, u(n), basefmt=' '); axes[0,0].set(xlim=[-10.5,10.5], title='\$u[n]\$')
axes[0,1].stem(n, ueven, basefmt=' '); axes[0,1].set(xlim=[-10.5,10.5], title='\$u_e[n]\$', xlabel='n')
axes[0,2].stem(n, uodd, basefmt=' '); axes[0,2].set(xlim=[-10.5,10.5], title='\$u_o[n]\$', xlabel='n')
axes[1,0].stem(n, x1(n), basefmt=' '); axes[1,0].set(xlim=[-10.5,10.5], title='\$x1[n]\$')
axes[1,1].stem(n, x1_even, basefmt=' '); axes[1,1].set(xlim=[-10.5,10.5], title='\$x1_e[n]\$', xlabel='n')
axes[1,2].stem(n, x1_odd, basefmt=' '); axes[1,2].set(xlim=[-10.5,10.5], title='\$x1_o[n]\$', xlabel='n')
axes[2,0].stem(n, x2(n), basefmt=' '); axes[2,0].set(xlim=[-10.5,10.5], title='\$x2[n]\$')
axes[2,1].stem(n, x2_even, basefmt=' '); axes[2,1].set(xlim=[-10.5,10.5], title='\$x2_e[n]\$', xlabel='n')
axes[2,2].stem(n, x2_odd, basefmt=' '); axes[2,2].set(xlim=[-10.5,10.5], title='\$x2_o[n]\$', xlabel='n')

fig.tight_layout()
```
