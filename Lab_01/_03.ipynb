```python
import numpy as np
import matplotlib.pyplot as plt
# 그림림 6.12
# x1 = sin(8pin/7); x2 = sin(10pin/4);
nmin,nmax = 0,240
n = np.arange(nmin,nmax+1)
x1= np.sin(8*np.pi*n/7)
x2= np.sin(10*np.pi*n/4)


fig,ax=plt.subplots(3,1,figsize=(7,4),dpi=50)
ax[0].stem(n,x1);
ax[1].stem(n,x2);
ax[2].stem(n,x1+x2);
```
