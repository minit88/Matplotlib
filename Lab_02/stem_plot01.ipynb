```python

import numpy as np
import matplotlib.pyplot as plt
n = np.arange(-5, 15+1)
# generate signals
imp = lambda n: (n==0)*(n==np.round(n)).astype(float)# impulse
u = lambda n: (n>=0)*(n==np.round(n)).astype(float)# step
x1 = np.exp(-0.5*n)*u(n)# exponential
x2 = np.cos(2*np.pi*n/12)# sinusoid
# stem-plot of signal
plt.figure(figsize=(6,4), dpi=100)
plt.subplot(1,3,1); plt.stem(n, imp(n)-2*imp(n-2)); plt.title('x1[n]')
plt.subplot(1,3,2); plt.stem(n, u(n)-u(n-5)); plt.title('x2[n]')
plt.subplot(1,3,3); plt.stem(n, ((0.8)**n)*(u(n)-u(n-10))); plt.title('x3[n]'); plt.xlabel('n')
plt.tight_layout();
```
