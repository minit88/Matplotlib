```python
import numpy as np
import matplotlib.pyplot as plt
h = np.array([0.5,1,0.5]); nh=np.arange(-1,1+1)
x= np.array([1,0,2,0,1,0,-1,0,-2,0,-1.2,0,1.2,0,2.4,0,1.5]) ; nx = np.arange(0,len(x))
y=np.convolve(x,h); ny=np.arange(min(nh)+min(nx),min(nh)+min(nx)+len(y))
flg,ax = plt.subplots(2,2,figsize=(10,4))
ax[0,0].stem(nh,h); ax[0,1].stem(nx,x); ax[1,1].stem(ny,y);
ax[0,1].set_xlim(min(ny),max(ny)); ax[1,1].set_xlim(min(ny),max(ny))
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(10); N = 1000;
x = np.random.rand(N); y = 2 * np.random.randn(N) + 3
fig, ax = plt.subplots(2,2, figsize=(6,4), dpi=50);
ax[0,0].plot(x); ax[0,1].plot(y)
ax[1,0].hist(x, bins=10, histtype='step');
ax[1,1].hist(y, bins=20);
print(np.mean(x), np.mean(y))
print(np.std(x), np.std(y))
```

# **FIR**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
N = 50; np.random.seed(10); n = np.arange(N); M = 5
s = 10*np.cos(2*np.pi*n/24); v = 2*np.random.randn(N);
x = s + v; h = np.ones(M)/M; 
# y = signal.lfilter(h, 1, x) # IIR y = np.convolve(x, h) # FIR
b = np.array([1/M, 0, 0, 0, 0, -1/M]); a = np.array([1, -1])
y = signal.lfilter(b, a, x)
fig, ax = plt.subplots(2, 1, figsize=(8,4), dpi=150)
ax[0].stem(x); ax[1].stem(y[M//2:]); ax[0].plot(s, 'r'); ax[1].plot(s, 'r');
```
