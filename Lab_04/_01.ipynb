```python
# 주파수 분석

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from numpy.fft import fft, fftshift

def ctft(x, tp):
  N = 8*len(x)
  X = fftshift(fft(x,N)) * tp
  f = np.linspace(-0.5/tp, 0.5/tp, len(X), endpoint=False)
  return X, f

N = 10; h = 1.5*(-0.5)**np.arange(N); fs = 2; X1, f = ctft(h, 1/fs); 
h2 = 1.5*(0.5)**np.arange(N); X2, f = ctft(h2, 1/fs); 
fig, ax = plt.subplots(2,2, figsize=(6,2), dpi=150)
ax[0,0].stem(h); ax[0,1].plot(f, np.abs(X1));
ax[1,0].stem(h2); ax[1,1].plot(f, np.abs(X2));
```
