```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from numpy.fft import fft,fftshift

x=np.array([1,2,-2,4]); h = np.array([1,-1,2])
#     1 -1 2
#        1 -1 2
#           1 -1 2
#              1 -1 2
#                 1 -1 2
#                   1 -1 2 
#                 
y = np.convolve(x,h)
rxh = np.correlate(x,h,'full')
yy= np.convolve(x,np.flip(h)) # 뒤집음 상관함수랑 같음
rxh2 = np.correlate(x,np.flip(h),'full')
print(y,rxh,yy,rxh2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from numpy.fft import fft,fftshift

fs=100; nTs = np.arange(0,1,1/fs)
f=10; x =  np.sin(2*np.pi*f*nTs) + 0.5*np.random.randn(len(nTs))
rxx = np.correlate(x,x,'full')
fig,ax =plt.subplots(2,1) 
ax[0].stem(x); ax[0].set(xlim=[0,20])
ax[1].stem(rxx) ; ax[1].set(xlim=[len(nTs)-15,len(nTs)+15])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from numpy.fft import fft,fftshift

delay = 500;
fs=100; nTs = np.arange(0,1,1/fs)
t=np.linspace(0,10,fs)
x=signal.chirp(t,f0=1,f1=5,t1=10,method='linear') # chirp 사용함
#x =  np.random.randn(len(nTs))
n=0.25*np.random.randn(len(x)+delay);
y=0.75*np.concatenate((np.zeros(delay),x))+n

rxx = np.correlate(x,y,'full')
fig,ax =plt.subplots(2,2) 
ax[0,0].plot(x); ax[0,1].plot(y)
ax[1,0].plot(rxx) 
```
