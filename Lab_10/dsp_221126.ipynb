```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from numpy.fft import fft,fftshift,ifft

#원형 컨벌루션
x1 =np.array([1,1,0,1]); x2=np.array([1,0,1,1])
X1= fft(x1);X2 = fft(x2)
y1 = ifft(X1 * X2)
print(np.real(y1))

#일반 컨벌루션
y=np.convolve(x1,x2); print(y)
x11 =np.concatenate((x1,np.zeros(len(x2)-1)))
x21 =np.concatenate((x2,np.zeros(len(x1)-1)))
X11=fft(x11);X21=fft(x21)

#일반 컨벌루션
y21 = ifft(X11*X21);print(np.real(y21))
```
