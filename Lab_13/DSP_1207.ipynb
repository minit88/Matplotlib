```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

# **IIR FILTER**

```python
b,a = signal.iirfilter(6,[2*np.pi*500],rp=2,rs=60, btype='low',
                       analog=True, ftype='cheby1')
w, h = signal.freqs(b, a,1000)

fig, ax = plt.subplots(1)

ax.plot(w/(2*np.pi), 20*np.log10(np.maximum(abs(h),1e-5)))
ax.axis((10, 2000, -100, 10))
ax.grid()
plt.show()
```
